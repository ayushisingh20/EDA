## Exploratory Data Analysis

GENDER

```python
ax =sns.countplot(x='Gender',data=df)
```

```python
print(ax)
```

```python
ax =sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Analysis : Female count more than male count
```

```python
df.groupby(['Gender'],as_index =False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
# Analysis : Female count as well as amount is more
```

```python
sales_gen = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by = 'Amount',ascending=False)
```

```python
sales_gen.head()
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
# Conclusion : From the above graphs we can see that most fo the buyers are women and purchasing power of women s greater than men
```

AGE

```python
sns.countplot(data=df,x='Age Group')
```

```python
sns.countplot(data=df,x='Age Group',hue='Gender')
```

```python
ax = sns.countplot(data=df,x='Age Group',hue ='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data= sales_age)
```

```python
sales_age.head()
```

```python
# From the above graphs we can see that the most buyers are of age group between 26-35 years female
```

STATE

```python
# Total no. of orders from top ten states

sales_state= df. groupby(['State'],as_index= False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sales_state.head()
```

```python
sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_state= df. groupby(['State'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

```python
# from above graphs we can see that most of the orders are from Uttar Pradesh, Maharashtra and karnataka respectively
```

MARITAL STATUS

```python
sns.set(rc = {'figure.figsize':(5,4)})
ax = sns.countplot(data = df, x ='Marital_Status')
```

```python
sales_status = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sales_status.head()
```

```python
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_status,x='Marital_Status',y='Amount',hue='Gender')
```

```python
# from the above graphs we can see that most of the buyers are married and they have high purchasing power
```

OCCUPATION

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df,x='Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state= df. groupby(['Occupation'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Occupation',y='Amount')
```

```python
# from the above graps we can see that most of the buyers are working in IT, Aviation and Healthcare Sector
```

PRODUCT CATEGORY

```python
sns.set(rc={'figure.figsize':(25,5)})
ax=sns.countplot(data=df,x='Product_Category')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state= df. groupby(['Product_Category'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Product_Category',y='Amount')
```

```python
# from the above graphs we can see that the most sold products are from food, clothing and electronics
```

PRODUCT ID

```python
sales_state= df. groupby(['Product_ID'],as_index= False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc = {'figure.figsize':(20,5)})
sns.barplot(data=sales_state,x='Product_ID',y='Orders')
```

Married women age group 26-35 years from UP, Maharashtra and Karnata working in IT, healthcare, Aviation are more likely to buy products from Food, Clothing and Electronics category
